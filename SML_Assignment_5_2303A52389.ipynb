```python
# reading the data with pandas
d1=pd.read_csv('/content/diabetes_data_upload.csv')
d1
```

```python
# check for null values
(d1.isnull())
```

```python
# If null values are there fill null values with 0
d1.fillna(0,inplace=True)
d1
```

```python
columns = ['Polyuria', 'Polydipsia', 'sudden weight loss', 'weakness', 'Polyphagia','Genital thrush', 'visual blurring', 'Itching', 'Irritability','delayed healing', 'partial paresis', 'muscle stiffness','Alopecia', 'Obesity']

d1[columns] = d1[columns].replace({'Yes': 1, 'No': 0})
d1['class'] = d1['class'].replace({'Positive': 1, 'Negative': 0})
d1['Gender'] = d1['Gender'].replace({'Male': 1, 'Female': 0})
d1
# here i have converted  the String "Yes" to 1 & "No" to 0.
# for class column "Positive" to 1 & "Negitive" to 0.
# for Gender column "Male" to 1 & "Female" to 0.
```

```python
hh = d1.drop('Gender', axis=1)
hh
```

```python
y1=d1['class']
y1
```

```python
#Features
x1=d1.drop('class',axis=1)
x1
```

```python
dd1=(x1-x1.min())/(x1.max()-x1.min())
dd1
```

```python
# graph b/w features and target
title = ['Polyuria', 'Polydipsia', 'sudden weight loss', 'weakness', 'Polyphagia','Genital thrush', 'visual blurring', 'Itching', 'Irritability','delayed healing', 'partial paresis', 'muscle stiffness','Alopecia', 'Obesity']

for col in title:
    x = d1[col]
    y = d1['class']
    plt.bar(x, y,edgecolor='k')
    plt.title(f'{col} vs class')
    plt.xlabel(f'Features  _ {col}')
    plt.ylabel('Target-class')
    plt.show()
```
